```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./hospital_mental(st).xlsx')
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
df
```

```python
ndf = pd.read_excel('./public_mental(st).xlsx')
ndf.drop(['Unnamed: 0'], axis = 1, inplace = True)
ndf
```

```python
result = pd.concat([df, ndf])
result
```

```python
result.drop_duplicates(subset = 'corp_name', keep='first', inplace=True)
result.index = range(len(result))
result
```

```python
result01 = result[result['address'].isnull()]
result01.index = range(len(result01))
df_hospital = result01.drop(['address'],axis=1)
df_hospital
```

```python
result02 = result[result['longitude'].isnull()]
result02.index = range(len(result02))
result02
df_mental = result02.drop(['longitude', 'latitude', 'treatment_subject'], axis = 1)
df_mental
```

```python
df_mental.to_excel('./public_mental(000).xlsx')
df_hospital.to_excel('./hospital_mental(000).xlsx')
```
