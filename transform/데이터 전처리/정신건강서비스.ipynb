```python
#기능1. json파일에서 정보 불러오기

with open('/Users/wooseongkyun/Datascience/프로젝트1/데이터정제/전국_정신건강관련기관_위치정보.json','r',encoding='utf-8') as f:
    json_data = json.load(f)
# with open(file_path, 'r') as j:
#     contents = json.loads(j.read())

#공공/민간 구분 기관 구분
public_or_private=[]
categories=[]
#상담소/민간의료기관/공공의료기관으로 구분
#상담소
counseling_names=[]
counseling_address=[]
counseling_latitude=[]
counseling_longitude=[]
counseling_phone_numbers=[]
counseling_website_urls=[]
#공공의료기관
public_names=[]
public_address=[]
public_latitude=[]
public_longitude=[]
public_phone_numbers=[]
public_website_urls=[]
#민간의료기관
private_names=[]
private_address=[]
private_latitude=[]
private_longitude=[]
private_phone_numbers=[]
private_website_urls=[]

#json에서 정보 불러오기
for i in range(len(json_data['data'])):
    public_or_private.append(json_data['data'][i]['공공/민간'])
    categories.append(json_data['data'][i]['기관구분'])
    if json_data['data'][i]['기관구분']=='상담소':
        counseling_names.append(json_data['data'][i]['기관명'])
        counseling_address.append(json_data['data'][i]['주소'])
        counseling_latitude.append(json_data['data'][i]['위도'])
        counseling_longitude.append(json_data['data'][i]['경도'])
        counseling_phone_numbers.append((json_data['data'][i]['전화번호']))
        counseling_website_urls.append(json_data['data'][i]['홈페이지'])
    elif json_data['data'][i]['공공/민간']=='공공':
        public_names.append(json_data['data'][i]['기관명'])
        public_address.append(json_data['data'][i]['주소'])
        public_latitude.append(json_data['data'][i]['위도'])
        public_longitude.append(json_data['data'][i]['경도'])
        public_phone_numbers.append((json_data['data'][i]['전화번호']))
        public_website_urls.append(json_data['data'][i]['홈페이지'])
    else:
        private_names.append(json_data['data'][i]['기관명'])
        private_address.append(json_data['data'][i]['주소'])
        private_latitude.append(json_data['data'][i]['위도'])
        private_longitude.append(json_data['data'][i]['경도'])
        private_phone_numbers.append((json_data['data'][i]['전화번호']))
        private_website_urls.append(json_data['data'][i]['홈페이지'])


print(public_names)
```

```python
#맵 만들기
time_start= time()

my_location=[37.55, 126.98]
my_map = folium.Map(location = my_location, zoom_start = 12,
                                tiles = 'OpenStreetMap',control_scale=True)

#서클 표시 Circle radius는 meter 기준, CircleMarker는 pixel 기준
folium.Circle(
    my_location,
    radius=5000,
    tooltip=folium.Tooltip,
    color='#ffffgg',
    fill_color='#3186cc',
    popup='내 중심 반경 5km',
).add_to(my_map)


# #선택 카테고리 추가하기
groups=folium.FeatureGroup(name='진료/상담 기관')
my_map.add_child(groups)
counseling_group=FeatureGroupSubGroup(groups,name='상담소')
my_map.add_child(counseling_group)
public_group=FeatureGroupSubGroup(groups,name='공공 의료기관(보건소/정신건강보건센터/대학병원 외)')
my_map.add_child(public_group)
private_group=FeatureGroupSubGroup(groups,name='민간 의료기관')
my_map.add_child(private_group)

#MarkerCluster함수를 통해 특정 수 이상의 마커는 뭉쳐서 보이도록 표시함

#상담소 마커
MC_conseling= MarkerCluster()
for name, lat, long, address_ in zip(counseling_names,counseling_latitude,counseling_longitude,counseling_address):
    text=f"""<h4 align='center'>{name}</h4><p align='center'>{address_}</p>"""
    iframe = folium.IFrame(text, width=100, height=100)
    popup = folium.Popup(iframe, max_width=3000)
    MC_conseling.add_child(
        folium.Marker([lat, long], popup = popup)).add_to(counseling_group)

#공공 의료기관 마커
MC_public=MarkerCluster()
for name, lat, long, address_ in zip(public_names,public_latitude,public_longitude,public_address):
    text=f"""<h4 align='center'>{name}</h4><p align='center'>{address_}</p>"""
    iframe = folium.IFrame(text, width=100, height=100)
    popup = folium.Popup(iframe, max_width=3000)
    MC_public.add_child(
        folium.Marker([lat, long], popup = popup)).add_to(public_group)

#민간 의료기관 마커
MC_private=MarkerCluster()
for name, lat, long, address_ in zip(private_names,private_latitude,private_longitude,private_address):
    text=f"""<h4 align='center'>{name}</h4><p align='center'>{address_}</p>"""
    iframe = folium.IFrame(text, width=100, height=100)
    popup = folium.Popup(iframe, max_width=3000)
    MC_private.add_child(
        folium.Marker([lat, long], popup = popup)).add_to(private_group)
folium.LayerControl(position='bottomright',collapsed=False).add_to(my_map)

# 지도 저장(seoul_colleges.html)
my_map.save('./mental_service.html')

time_end=time()
print(time_end-time_start)
```
