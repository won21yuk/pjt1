```python
import pandas as pd
import numpy as np
import pyproj
import folium
```

```python
df = pd.read_csv('./hostipal.csv', encoding='cp949', usecols=['사업장명', '진료과목내용명', '좌표정보(x)', '좌표정보(y)'])

df['좌표정보(x)'] = pd.to_numeric(df['좌표정보(x)'], errors='coerce')
df['좌표정보(y)'] = pd.to_numeric(df['좌표정보(y)'], errors='coerce')

df = df.dropna()

df = df[df['진료과목내용명'].str.contains('정신건강의학과')]
df.columns = ['corp_name', 'longitude', 'latitude', 'treatment_subject']
df.index = range(len(df))
df
```

```python
ndf = pd.read_csv('./mental_health.csv', encoding='cp949', usecols=['기관명', '주소'])

ndf = ndf.dropna()

ndf.index = range(len(ndf))
ndf
ndf.columns = ['corp_name','address']
ndf
```

```python
result = pd.concat([df, ndf])
result
```

```python
result.drop_duplicates(subset = 'corp_name', keep='first', inplace=True)
result
result.index = range(len(result))
result
```

```python
# result.to_excel('./result.xlsx')

result01 = result[result['address'].isnull()]
result01.index = range(len(result01))
df_hospital = result01.drop(['address'],axis=1)
df_hospital
```

```python
def project_array(coord, p1_type, p2_type):
    """
    좌표계 변환 함수
    - coord: x, y 좌표 정보가 담긴 NumPy Array
    - p1_type: 입력 좌표계 정보 ex) epsg:5179
    - p2_type: 출력 좌표계 정보 ex) epsg:4326
    """
    p1 = pyproj.Proj(init=p1_type)
    p2 = pyproj.Proj(init=p2_type)
    fx, fy = pyproj.transform(p1, p2, coord[:, 0], coord[:, 1])
    return np.dstack([fx, fy])[0]
```

```python
df_array = df_hospital.drop(['corp_name', 'treatment_subject'], axis=1)
df_array
```

```python
coord = np.array(df_array)

coord
```

```python
p1_type = "epsg:2097"
p2_type = "epsg:4326"

result = project_array(coord, p1_type, p2_type)
result
```

```python
df_hospital['longitude'] = result[:, 0]
df_hospital['latitude'] = result[:, 1]

df_hospital
```
