```python
import json
import pandas as pd
import numpy as np

with open('./hospital_psy_hira.json', 'r', encoding='utf-8') as f:
    data_hira = json.load(f)
    
df_hira = pd.DataFrame(data_hira)

with open('./mental_care_final.json', 'r', encoding='utf-8') as d:
    data_care = json.load(d)

df_care = pd.DataFrame(data_care)
    
df_hira.rename(columns ={'longitude':'위도'}, inplace=True)
df_hira.rename(columns ={'latitude':'경도'}, inplace=True)

df = pd.concat([df_hira, df_care])
df.drop_duplicates(subset='기관명', keep='first', inplace=True, ignore_index=False)

df['위도'] = pd.to_numeric(df['위도'], errors='coerce')
df['경도'] = pd.to_numeric(df['경도'], errors='coerce')


df
```

```python
df.drop_duplicates(subset='주소', keep='first', inplace=True, ignore_index=False)
df.drop_duplicates(subset='위도', keep='first', inplace=True, ignore_index=False)
df.drop_duplicates(subset='경도', keep='first', inplace=True, ignore_index=False)

df.index = range(len(df))

df
```

```python
df.to_json('edition.json', force_ascii=False, orient = 'records', indent = 4)
```
