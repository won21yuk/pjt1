```python
import folium
import pandas as pd
```

```python
m_seoul = folium.Map(location=[37.536082, 126.983967], zoom_start=12)
m_seoul
```

```python
import json


with open('./hospital_psy_hira.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
    
print(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
df_seoul = df[df['주소'].str.contains('서울특별시')]
df_seoul.index = range(len(df_seoul))
df_seoul
```

```python
for name, lat, lng in zip(df_seoul.기관명, df_seoul.latitude, df_seoul.longitude):
    folium.Marker([lat,lng], popup=name).add_to(m_seoul)
    
m_seoul
```

```python
import math
def getLatLngBound(lat, lng):
  # 5km 구간
  lat_change = 5 / 111.2
  lng_change = abs(math.cos(lat * (math.pi / 180)))
  bounds = {
    "lat_min": lat - lat_change,
    "lng_min": lng - lng_change,
    "lat_max": lat + lat_change,
    "lng_max": lng + lng_change
  }
  return bounds
```

```python
plugins.LocateControl(locateOptions={maxZoom:10})add_to(m_seoul)
m_seoul
```
